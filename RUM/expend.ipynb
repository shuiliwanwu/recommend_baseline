```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('train1.tsv',sep='\t',header=None)
test=pd.read_csv('test1.tsv',sep='\t',header=None)
```

```python
train.columns=['user','item','pre']
test.columns=['user','item','pre']
```

```python
train['pre1']=train['pre'].map(lambda x:int(x[1:-1].split()[0]))
train['pre2']=train['pre'].map(lambda x:int(x[1:-1].split()[1]))
train['pre3']=train['pre'].map(lambda x:int(x[1:-1].split()[2]))
train['pre4']=train['pre'].map(lambda x:int(x[1:-1].split()[3]))
train['pre5']=train['pre'].map(lambda x:int(x[1:-1].split()[4]))
```

```python
test['pre1']=test['pre'].map(lambda x:int(x[1:-1].split()[0]))
test['pre2']=test['pre'].map(lambda x:int(x[1:-1].split()[1]))
test['pre3']=test['pre'].map(lambda x:int(x[1:-1].split()[2]))
test['pre4']=test['pre'].map(lambda x:int(x[1:-1].split()[3]))
test['pre5']=test['pre'].map(lambda x:int(x[1:-1].split()[4]))
```

```python
train.pop('pre')
test.pop('pre')
```

```python
train.to_csv('train2.tsv',index=None,header=None,sep='\t')
test.to_csv('test2.tsv',index=None,header=None,sep='\t')
```
