```python
import pandas as pd
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
train=pd.read_csv('train.tsv',header=None,sep='\t')
train.columns=['user','item']
```

```python
temp=pd.read_csv('itemmap.csv')
dict1={}
for i in range(len(temp)):
    zz=temp.iloc[i]
    dict1[zz['ItemId']]=zz['item_idx']
```

```python
train['item']=train['item'].agg(lambda x: dict1[x])
```

```python
le.fit(train['user'])
train['user']=le.transform(train['user'])
```

```python
train=train.reset_index()
index=train.groupby('user')['index'].agg(max).reset_index()
index=index['index']
user=[]
item=[]
pre=[]
for t in range(len(index)):
    if t==0:
        pre_index=0
    else:
        pre_index=index[t-1]
    for i in range(pre_index+5,index[t]+1):
        user.append(train.iloc[i]['user'])
        item.append(train.iloc[i]['item'])
        pre.append(train.iloc[i-5:i]['item'].values)        
```

```python
train_new=pd.DataFrame()
train_new['user']=user
train_new['item']=item
train_new['pre']=pre
```

```python
test=pd.read_csv('test.tsv',header=None,sep='\t')
test.columns=['user','item']
test['item']=test['item'].agg(lambda x: dict1[x])
test['user']=le.transform(test['user'])
test=test.reset_index()
index=test.groupby('user')['index'].agg(max).reset_index()
index=index['index']
test1=test.iloc[index]
test1.pop('index')
pre1=[]
for t in index:
    pre1.append(test.iloc[t-5:t]['item'].values)
    

test1['pre']=pre1

test1=test1.reset_index(drop=True)
```

```python
train_new.to_csv('train1.tsv',index=None,header=None,sep='\t')
test1.to_csv('test1.tsv',index=None,header=None,sep='\t')
```
